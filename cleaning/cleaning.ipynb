```python
match = "biological sex means"
rx_to_first = r'^.*?{}'.format(re.escape(match))
print(re.sub(rx_to_first, f'The term {match}', definition))
```

```python
matches = df['Match']
defis = df['Definition']
```

```python
regularized = []
for match, defi in zip(matches, defis):
    rx_to_first = r'^.*?{}'.format(re.escape(match))
    regularized.append(re.sub(rx_to_first, f'The term {match}', defi))
```

## removing characters regex
- remove everything after a period. 
- remove single and double characters (both numbers and alphabetic characters) within quotes
- remove double hyphens

```python
for item in regularized:
    print(item)
```

```python
# Regex to find everything after the period but keeping the period
pattern = r"\.(.*)"

# Replace the found pattern with just a period
# cleaned_text = re.sub(pattern, ".", text)

results = []
for item in regularized:
    cleaned = re.sub(pattern, '.', item)
    results.append(cleaned)
```

```python
results
```

```python
# Sample text
text = "Here are examples: (a), (1), (ab), (12), and (abc) should not be removed."

# Regex to find single or double characters within parentheses
pattern = r"\([a-zA-Z0-9]{1,2}\)"

# Replace the found patterns with an empty string (removing them)
cleaned_text = re.sub(pattern, "", text)

print(cleaned_text)
```

```python
# Regex to find single or double characters within parentheses
pattern = r"\([a-zA-Z0-9]{1,2}\)"

no_paren = []
for item in results:
    cleaned = re.sub(pattern, '.', item)
    no_paren.append(cleaned)
```

```python
no_paren
```

```python
with open('test_no_parentheses.txt', 'w') as f:
    for item in no_paren:
        f.write(item+"\n")
```
